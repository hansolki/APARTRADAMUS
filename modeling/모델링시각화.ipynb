```python
# df 원본 불러오기
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/ML_history_ver.5.5_FilteredDataset_AutoML_XGBoost_LGBM의 사본 - History.csv', index_col=0)
df.head()
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.iloc[2:87][['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 9']]
df
```

```python
df.columns=['Time', 'Train_R2', 'Test_R2']
df
```

```python
df = df.astype(float)
df.info()
```

```python
df = df.round(3)
df
```

```python
df2 = df.iloc[:10]
df2
```

```python
from matplotlib.ticker import MultipleLocator
```

```python
# 첫 번째 축에 대한 그래프
fig, ax1 = plt.subplots(figsize=(9, 4))
color = 'tab:blue'
ax1.set_xlabel('Time')
ax1.set_ylabel('Train_R2', color=color)
ax1.plot(df2['Time'], df2['Train_R2'], color=color, label='Train_R2', marker='o')
ax1.tick_params(axis='y', labelcolor=color)

# 두 번째 축 생성
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Test_R2', color=color)
ax2.plot(df2['Time'], df2['Test_R2'], color=color, label='Test_R2', marker='o')
ax2.tick_params(axis='y', labelcolor=color)

# 왼쪽과 오른쪽 축의 범위를 동일하게 설정
min_value = min(df2['Train_R2'].min(), df2['Test_R2'].min())
max_value = max(df2['Train_R2'].max(), df2['Test_R2'].max())
ax1.set_ylim(0.65, 1.00)
ax2.set_ylim(0.65, 1.00)

# x축 눈금을 60의 배수로 설정
major_locator = MultipleLocator(60)
ax1.xaxis.set_major_locator(major_locator)
ax2.xaxis.set_major_locator(major_locator)

# ax2의 최댓값에 대한 표시 (노란색 마커)
max_value_index_ax2 = df2['Test_R2'].idxmax()
max_value_x_ax2 = df2.loc[max_value_index_ax2, 'Time']
max_value_y_ax2 = df2.loc[max_value_index_ax2, 'Test_R2']
ax2.scatter(max_value_x_ax2, max_value_y_ax2, color='yellow', s=150, edgecolors='black', linewidth=2, alpha=1, label='Max Y2')

# 최댓값에 대응하는 ax1의 값을 찾아 표시
max_value_index_ax1 = df2['Time'].idxmax()
max_value_y_ax1 = df2.loc[max_value_index_ax1, 'Train_R2']
ax1.scatter(max_value_x_ax2, max_value_y_ax1, color='yellow', s=150, edgecolors='black', linewidth=2, alpha=1, label='Max Y1 Corresponding')

# 최댓값에 대한 숫자 표시 (마커 아래로 내리기)
ax2.annotate(f'{max_value_y_ax2:.3f}', (max_value_x_ax2, max_value_y_ax2),
             textcoords="offset points", xytext=(0, -20), ha='center')
ax1.annotate(f'{max_value_y_ax1:.3f}', (max_value_x_ax2, max_value_y_ax1),
             textcoords="offset points", xytext=(0, -20), ha='center')

plt.title('2nd Modeling: AutoML(FLAML)')
plt.show()
```
