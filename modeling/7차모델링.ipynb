```python
# 상관관계 분석
df_corr = df.corr()
df_corr['거래금액(만원)'] = abs(df_corr['거래금액(만원)'])
df_corr = df_corr.sort_values('거래금액(만원)', ascending=False)
df_corr.iloc[:,-1]
```

```python
# df 컬럼명 전처리
df.columns = df.columns.str.replace(r'\([^)]*\)', '', regex=True)
df.columns = df.columns.str.replace(r' ', '_', regex=True)
df.columns


# MinMaxScaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = df.select_dtypes(exclude='object').columns
cols = cols.drop('거래금액')
df[cols] = scaler.fit_transform(df[cols])
df.head()


# Label Encoding
from sklearn.preprocessing import LabelEncoder
cols = ['지역구', '동']
for col in cols:
  le = LabelEncoder()
  df[col] = le.fit_transform(df[col])
df.head()


# train test 나누기
X_train = df[df['계약년']<=0.6].drop('거래금액', axis=1)
X_test = df[df['계약년']>=0.6].drop('거래금액', axis=1)
y_train = df[df['계약년']<0.6]['거래금액']
y_test = df[df['계약년']>=0.6]['거래금액']
X_train.shape, X_test.shape, y_train.shape, y_test.shape


# AutoML
from flaml import AutoML
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

automl = AutoML()
times = [i for i in range(60, 601, 60)] # 1분~10분까지
estimators = [['lgbm'], ['xgboost']] # 집중적으로 돌려볼 모델

for estimator in estimators:
  for time in times:

    automl.fit(X_train=X_train, y_train=y_train, task="regression", estimator_list=estimator, time_budget=time, verbose=False)
    print('time:', time)
    print('Best ML model:', automl.model)
    print('Best hyperparameter config:', automl.best_config)
    print('Best accuracy on validation data: {0:.4g}'.format(1-automl.best_loss))
    print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))

    y_pred = automl.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print('Mean Squared Error (MSE) =', mse)
    print('Mean Absolute Error (MAE) =', mae)
    print('R-squared (R2) =', r2)

    ft = pd.Series(automl.feature_importances_,
          index=automl.feature_names_in_)\
    .sort_values(ascending=False)
    print(ft)

    print('\n')
```
