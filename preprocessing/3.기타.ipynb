```python
# 파일 경로 설정
file_path = '/content/drive/MyDrive/2조/Database/독립변수/8.기타/5대+범죄+발생현황(2015~2022).csv'

# 파일 읽어오기
import pandas as pd
df = pd.read_csv(file_path)
# df.head(20)

df = df[['자치구별(2)', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]
df

df.columns = ['지역구', '2015 범죄건수', '2016 범죄건수', '2017 범죄건수', '2018 범죄건수', '2019 범죄건수', '2020 범죄건수', '2021 범죄건수', '2022 범죄건수']
df

df = df.drop([0,1,2,3], axis=0).reset_index(drop=True)
df
```

```python
def process_row(row):
    data = {}
    for i in row.index[1:]:
        v = data.get(i[:4], {})
        v[i[5:]] = row[i]
        data[i[:4]] = v

    table = pd.DataFrame(data).T.reset_index()
    table['지역구'] = row['지역구']
    table.columns = ['연도', '범죄건수', '지역구']
    table = table[['지역구', '연도', '범죄건수']]
    return table

tables = [process_row(row) for _, row in df.iterrows()]
df_new = pd.concat(tables, ignore_index=True)
df_new.head(30)

df_new.to_csv('연도별 지역구별 범죄건수.csv')
```

```python
# 파일 경로 설정
file_path = '/content/drive/MyDrive/2조/Database/독립변수/8.기타/인플루엔자_예방접종률(2010~2022).csv'

# 파일 읽어오기
import pandas as pd
df = pd.read_csv(file_path)
df = df[df['시군구별(1)']=='서울특별시']
df

df = df[['시군구별(2)', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]
df.head(35)

df.columns = ['지역구', '2015 인플루엔자 접종률', '2016 인플루엔자 접종률', '2017 인플루엔자 접종률', '2018 인플루엔자 접종률', '2019 인플루엔자 접종률', '2020 인플루엔자 접종률', '2021 인플루엔자 접종률', '2022 인플루엔자 접종률']
df

df = df.drop([0], axis=0).reset_index(drop=True)
df
```

```python
def process_row(row):
    data = {}
    for i in row.index[1:]:
        v = data.get(i[:4], {})
        v[i[5:]] = row[i]
        data[i[:4]] = v

    table = pd.DataFrame(data).T.reset_index()
    table['지역구'] = row['지역구']
    table.columns = ['연도', '인플루엔자 접종률', '지역구']
    table = table[['지역구', '연도', '인플루엔자 접종률']]
    return table

tables = [process_row(row) for _, row in df.iterrows()]
df_new = pd.concat(tables, ignore_index=True)
df_new

df_new.to_csv('연도별 지역구별 인플루엔자 접종률.csv')
```
