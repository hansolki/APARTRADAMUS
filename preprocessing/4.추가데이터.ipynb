```python
# 파일 경로 설정
file_path = '/content/drive/MyDrive/2조/Database/독립변수/9.추가/독거노인가구비율_시도_시_군_구__20231222105622.csv'

# 파일 읽어오기
import pandas as pd
df2 = pd.read_csv(file_path)
df2.head(20)

df2 = df2[['행정구역별', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]
df2

df2.columns = ['지역구', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
df2

df2 = df2.drop([0,1], axis=0).reset_index(drop=True)
df2
```

```python
#df1에 '공공체육시설'이라는 칼럼을 추가하고, df2의 연도별, 지역구 조건에 맞는 데이터를 해당 칼럼에 넣는 함수
def add_sports_facilities_column(df1, df2):
    # df2의 연도 열 이름을 리스트로 변환
    years = df2.columns[1:]

    # df2를 녹여서 각 지역구의 연도별 공공체육시설 수를 단일 열로 만듬
    df2_melted = df2.melt(id_vars=['지역구'], value_vars=years, var_name='계약년', value_name='독거노인가구비율(%)')
    df2_melted['계약년'] = df2_melted['계약년'].astype(int)  # 계약년 형변환

    # df1과 df2_melted를 '지역구'와 '계약년'을 기준으로 병합
    df1_merged = pd.merge(df1, df2_melted, on=['지역구', '계약년'], how='left')

    return df1_merged

#함수 실행
df1 = add_sports_facilities_column(df1, df2)

#결과의 처음 몇 행 확인
df1.head()
```
