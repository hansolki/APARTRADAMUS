```python
# for row in df.iterrows():
tables = []
for idx,row in df.iterrows():
  # print(row)
  # print(type(row))
  # print(row['지역구'])
  # print(row.index)
  data = {}
  for i in row.index[1:]:
    # print(i[:4])
    # print(i[5:])
    v = data.get(i[:4], {})
    v[i[5:]] = row[i]
    data[i[:4]] = v

  table = pd.DataFrame(data).T.reset_index()
  table['지역구'] = row['지역구']
  table.columns = ['연도', '초등학교 수', '고등학교 수', '학원 수', '지역구']
  table = table[['지역구', '연도', '초등학교 수', '고등학교 수', '학원 수']]
  # table.T
  # print(table)
  tables.append(table)

  # break
df_new = pd.concat(tables, ignore_index=True)
df_new.head(30)

df_new.to_csv('초+고+학원.csv')
```
